```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Akshay\Downloads\lifeexpPCA.csv")
df.head()
```

```python
df.shape
```

```python
X_train=df.iloc[:,2:20]
y_train=df.iloc[:,0:2]
```

```python
X_train.head(5)
```

```python
y_train.head(5)
```

```python
from sklearn.decomposition import PCA 
model = PCA(n_components = 5)
principalComponents= model.fit(X_train)
```

```python
pca=list(np.round(principalComponents.explained_variance_,2)) 
pca
```

```python
principalComponents.components_ 
```

```python
transformed_data = principalComponents.fit_transform(X_train)
transformed_data
```

```python
print("original shape:   ", X_train.shape)
print("transformed shape:", transformed_data.shape)
```

```python
principalDf = pd.DataFrame(data = transformed_data
             , columns = ['Data PC1', 'Data PC2','Data PC3','Data PC4','Data PC5'])
principalDf.head()
```

```python
df_final=pd.concat((principalDf,y_train),axis=1)
df_final.head(5)
```

```python
df_final.to_csv(r"C:\Users\Akshay\Downloads\df_finalMLR.csv",index=False)
```
